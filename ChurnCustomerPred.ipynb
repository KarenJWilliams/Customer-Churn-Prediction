```python
#1__Load Libraries and Dataset
# Import all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Display settings for better readability
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid", palette="muted")

# Load dataset
data_path = r"C:\Users\User\.cache\kagglehub\datasets\rashadrmammadov\customer-churn-dataset\versions\1\customer_churn_data.csv"
df = pd.read_csv(data_path)

# Preview
print("âœ… Dataset Loaded Successfully!")
print(f"Shape: {df.shape}")
df.head()
```

```python
#2__Dataset Overview
# General information about the dataset
print("\n--- Dataset Info ---")
df.info()

# Basic statistics (numerical + categorical)
print("\n--- Numerical Summary ---")
display(df.describe())

print("\n--- Categorical Summary ---")
display(df.describe(include='object'))

# Missing values
print("\n--- Missing Values ---")
print(df.isnull().sum())

# Duplicates
print(f"\n--- Duplicates --- {df.duplicated().sum()} rows found")

# Unique value counts for each column
print("\n--- Unique Values per Column ---")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
#3__Data Cleaning (Detailed Column-wise)
# 1. Drop duplicates
df.drop_duplicates(inplace=True)

# 2. Handle missing values column-wise
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype == 'object':
            print(f"Filling missing values in {col} with mode: {df[col].mode()[0]}")
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            print(f"Filling missing values in {col} with median: {df[col].median()}")
            df[col].fillna(df[col].median(), inplace=True)

# 3. Convert data types if needed
# Example: If TotalCharges is stored as string
if 'TotalCharges' in df.columns and df['TotalCharges'].dtype == 'object':
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
    df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

# 4. Remove outliers using IQR method for numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]

# 5. Encode categorical variables
labelencoder = LabelEncoder()
for col in df.select_dtypes(include='object').columns:
    df[col] = labelencoder.fit_transform(df[col])

print("\nâœ… Data Cleaning Completed!")
print(f"New shape: {df.shape}")
df.head()
```

```python
#4__Detailed Data Analysis â€” Numerical Columns

# Identify numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure(figsize=(14,5))

    # Distribution plot
    plt.subplot(1,2,1)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)

    # Box plot to identify outliers
    plt.subplot(1,2,2)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()

    # Statistical info
    print(f"ðŸ“Š {col} Summary:")
    print(df[col].describe())
    print("-" * 80)
```

```python
#5__Detailed Data Analysis â€” Categorical Columns

# Identify categorical columns
cat_cols = [col for col in df.columns if df[col].nunique() < 20 and df[col].dtype != 'float64']

for col in cat_cols:
    plt.figure(figsize=(10,5))
    sns.countplot(x=col, data=df, palette='pastel')
    plt.title(f"Frequency Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

    # Percentage distribution
    print(f"ðŸ“Š {col} Value Counts (%):")
    print((df[col].value_counts(normalize=True) * 100).round(2))
    print("-" * 80)
```

```python
#6__Relationship Analysis (Bivariate Analysis)

target = 'Churn' if 'Churn' in df.columns else None

if target:
    # Numerical vs Target
    for col in num_cols:
        plt.figure(figsize=(6,4))
        sns.boxplot(x=target, y=col, data=df, palette='coolwarm')
        plt.title(f"{col} vs {target}")
        plt.show()

    # Categorical vs Target
    for col in cat_cols:
        if col != target:
            plt.figure(figsize=(8,4))
            sns.countplot(x=col, hue=target, data=df, palette='viridis')
            plt.title(f"{col} vs {target}")
            plt.xticks(rotation=30)
            plt.show()

    # Correlation Heatmap
    plt.figure(figsize=(12,8))
    sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="YlGnBu")
    plt.title("Correlation Heatmap")
    plt.show()
```

```python
#7__Statistical Relationships

from scipy.stats import chi2_contingency, pearsonr

print("\n--- Statistical Relationship Testing ---")

if target:
    for col in cat_cols:
        if col != target:
            contingency = pd.crosstab(df[col], df[target])
            chi2, p, dof, expected = chi2_contingency(contingency)
            print(f"{col}: Chi-Square p-value = {p:.4f}")

    for col in num_cols:
        if col != target:
            corr, pval = pearsonr(df[col], df[target])
            print(f"{col}: Pearson correlation with {target} = {corr:.3f}, p-value = {pval:.4f}")
```

```python
#1. Importing Required Libraries
# Machine Learning and Evaluation Libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve

# Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

print("âœ… Libraries successfully imported!")
```

```python
from IPython.display import display, Markdown

markdown_text = """
ðŸ§© 2. Feature Selection and Rationale
ðŸŽ¯ Objective:
To identify the most significant predictors of customer churn using statistical correlation, domain understanding, and data exploration results from the previous phase.

| Feature          | Type      | Justification                                                                 |
|------------------|-----------|-------------------------------------------------------------------------------|
| tenure           | Numeric  | Represents customer loyalty â€” shorter tenure indicates higher churn risk.     |
| MonthlyCharges   | Numeric  | High monthly bills often increase dissatisfaction and churn.                  |
| TotalCharges     | Numeric  | Indicates long-term engagement; low values correlate with early churn.       |
| Contract         | Categorical | Month-to-month contracts show higher churn compared to long-term ones.     |
| InternetService  | Categorical | Internet type can affect perceived service quality.                        |
| OnlineSecurity   | Categorical | Add-on services like online security reduce churn likelihood.               |
| TechSupport      | Categorical | Customers with technical support are less likely to leave.                  |
| PaymentMethod    | Categorical | Auto-payment methods often correlate with customer retention.               |
| SeniorCitizen    | Binary    | Provides demographic differentiation â€” may affect service preferences.       |"""

display(Markdown(markdown_text))
```

```python
# Select relevant features and target
selected_features = [
    'tenure', 'MonthlyCharges', 'TotalCharges',
    'Contract', 'InternetService', 'OnlineSecurity',
    'TechSupport', 'PaymentMethod', 'SeniorCitizen'
]

X = df[selected_features]
y = df['Churn']
```

```python
#3. Data Preparation
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# Feature scaling for numeric variables
scaler = StandardScaler()
X_train[['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(
    X_train[['tenure', 'MonthlyCharges', 'TotalCharges']]
)
X_test[['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.transform(
    X_test[['tenure', 'MonthlyCharges', 'TotalCharges']]
)

print("âœ… Data successfully split and scaled!")
print(f"Training data shape: {X_train.shape}")
```

```python
from IPython.display import display, Markdown

markdown_text = """
4. Model Building and Rationale
| Model                  | Type                     | Rationale                                                                 |
|------------------------|--------------------------|---------------------------------------------------------------------------|
| Logistic Regression     | Linear Classification     | A simple yet powerful baseline model that provides interpretable coefficients to understand how features affect churn. |
| Random Forest Classifier| Ensemble (Bagging)        | Reduces overfitting, handles nonlinear relationships, and provides feature importance scores. |
| XGBoost Classifier      | Ensemble (Boosting)       | High-performance algorithm capable of capturing complex patterns and handling data imbalance efficiently. |"""
display(Markdown(markdown_text))
```

```python
#Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)
```

```python
#Random Forest Classifier
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
#XGBoost Classifier
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

```python
#5. Model Evaluation Metrics
def evaluate_model(name, y_true, y_pred):
    print(f"ðŸ“Š Model: {name}")
    print("Accuracy:", round(accuracy_score(y_true, y_pred), 4))
    print("ROC-AUC:", round(roc_auc_score(y_true, y_pred), 4))
    print("\nClassification Report:\n", classification_report(y_true, y_pred))
    print("-"*80)

evaluate_model("Logistic Regression", y_test, y_pred_lr)
evaluate_model("Random Forest", y_test, y_pred_rf)
evaluate_model("XGBoost", y_test, y_pred_xgb)
```

```python
#6. Confusion Matrix Visualization
models = {
    "Logistic Regression": y_pred_lr,
    "Random Forest": y_pred_rf,
    "XGBoost": y_pred_xgb
}

plt.figure(figsize=(15,4))
for i, (name, pred) in enumerate(models.items()):
    plt.subplot(1,3,i+1)
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
#7. ROC Curve Analysis
plt.figure(figsize=(8,6))

y_prob_lr = log_reg.predict_proba(X_test)[:,1]
y_prob_rf = rf.predict_proba(X_test)[:,1]
y_prob_xgb = xgb.predict_proba(X_test)[:,1]

fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)

plt.plot(fpr_lr, tpr_lr, label='Logistic Regression')
plt.plot(fpr_rf, tpr_rf, label='Random Forest')
plt.plot(fpr_xgb, tpr_xgb, label='XGBoost')
plt.plot([0,1], [0,1], 'k--')

plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
""" PHASE3:MODEL OPTIMIZATION, EVALUATION AND DEPLOYEMENT
"""
```

```python
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from xgboost import XGBClassifier
import numpy as np

# Define numeric and categorical columns
numeric_features = ['tenure', 'MonthlyCharges', 'TotalCharges']
categorical_features = ['Contract', 'InternetService', 'OnlineSecurity',
                        'TechSupport', 'PaymentMethod', 'SeniorCitizen']

# Detect sklearn version and choose correct arg name
import sklearn
sk_version = tuple(map(int, sklearn.__version__.split(".")[:2]))

if sk_version >= (1, 2):
    categorical_transformer = Pipeline(
        steps=[('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))]
    )
else:
    categorical_transformer = Pipeline(
        steps=[('ohe', OneHotEncoder(handle_unknown='ignore', sparse=False))]
    )

# Build preprocessing pipeline
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# Full modeling pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(use_label_encoder=False,
                                 eval_metric='logloss',
                                 random_state=42))
])

# Stratified K-Fold Cross-Validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = ['accuracy', 'roc_auc', 'precision', 'recall', 'f1']

cv_results = cross_validate(pipeline, X, y, cv=cv, scoring=scoring, n_jobs=-1)
cv_summary = {m: round(cv_results[f'test_{m}'].mean(), 3)
              for m in ['accuracy', 'roc_auc', 'precision', 'recall', 'f1']}

print("ðŸ“Š Cross-Validation Performance Summary:")
for k, v in cv_summary.items():
    print(f"{k:<10}: {v}")
```

```python
# ===========================
# Phase 3.2 â€” Hyperparameter Optimization
# ===========================
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [3, 5, 7],
    'classifier__learning_rate': [0.01, 0.05, 0.1],
    'classifier__subsample': [0.7, 0.9, 1.0],
    'classifier__colsample_bytree': [0.7, 0.9, 1.0],
}

search = RandomizedSearchCV(
    pipeline,
    param_distributions=param_grid,
    scoring='roc_auc',
    cv=cv,
    n_iter=15,
    n_jobs=-1,
    verbose=1,
    random_state=42
)

search.fit(X, y)
print("\nðŸ† Best ROC-AUC:", round(search.best_score_, 4))
print("Best Parameters:", search.best_params_)
best_pipeline = search.best_estimator_
```

```python
# ===========================
# Phase 3.3 â€” Final Evaluation
# ===========================
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)

best_pipeline.fit(X_train, y_train)
y_pred = best_pipeline.predict(X_test)
y_prob = best_pipeline.predict_proba(X_test)[:, 1]

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_prob):.3f}")

# Confusion Matrix
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Greens')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc_score(y_test, y_prob):.3f})')
plt.plot([0,1], [0,1], linestyle='--', color='gray')
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
# ===========================
# Phase 3.4 â€” Model Explainability (SHAP)
# ===========================
import shap

model = best_pipeline.named_steps['classifier']
preproc = best_pipeline.named_steps['preprocessor']

X_train_trans = preproc.transform(X_train)

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_train_trans)

shap.summary_plot(shap_values, X_train_trans, feature_names=preproc.get_feature_names_out())
```
